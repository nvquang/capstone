```python
def position_predict(cand_id):
    
    company_df.loc[company_df['startYear'] == 'Present', 'startYear'] = 2017
    company_df.loc[company_df['endYear'] == 'Present', 'endYear'] = 2017
    
    e_1 = {}
    e_2 = {}
    
    companies = company_df[company_df.cand_id == cand_id]
    
    for index, row in companies.iterrows():
        country = row['country']
        
        if e_1.get(country) == None:
            e_1[country] = 0
        e_1.update({country: e_1.get(country) + 1})
        
        contient = contient_dictionary.get(country)
        if e_2.get(contient) == None:
            e_2[contient] = 0
        e_2.update({contient: e_2.get(contient) + 1})
    
    for e in e_1.keys():
        e_1.update({e: e_1.get(e)*1.0/len(companies)})
        
    for e in e_2.keys():
        e_2.update({e: e_2.get(e)*1.0/len(companies)})
        
    return (e_1, e_2)
```
